```python
unit_attack_dict = {
    "테란" : 10,
    "저그" : 12,
    "프로토스" :15
}

class Unit(object):
  total_unit_count = 0
  def __init__(self, name):
    self._name = name
    Unit.total_unit_count += 1
  def attack(self):
    return unit_attack_dict.get(self._name,None)
  def __del__(self):
    Unit.total_unit_count -= 1
```

```python
z1 = Unit("저그")
z2 = Unit("저그")
p1 = Unit("프로토스")
```

```python
print(id(z1))
print(id(z2))
print(id(p1))
```
