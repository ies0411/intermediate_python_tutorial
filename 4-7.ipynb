```python
from abc import *

class Animal(metaclass=ABCMeta):
    @abstractmethod
    def bark(self):
        print('animal : bark')

    @abstractmethod
    def walk(self):
        pass

    def sleep(self):
        print('animal : sleep')

class Human(Animal):
    def thinking(self):
        print('human : thinking')

class Dog(Animal):
    def walk(self):
        print('Dog : walk')

    def bark(self):
        print('Dog : bark')

# animal = Animal() # error
# human = Human() # error

dog = Dog()

dog.walk()
dog.bark()
dog.sleep()
```
