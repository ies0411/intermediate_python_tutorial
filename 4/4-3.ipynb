```python
class PropertySample:
  def __init__(self):
      self.x = 0
      self._z = 0
      self.__y = 0
  @property
  def y(self):
    return self.__y
  @y.setter
  def y(self, value):
    self.__y = value
  @y.deleter
  def y(self):
    del self.__y
```

```python
sample = PropertySample()
sample.y = 2
del sample.y
```
