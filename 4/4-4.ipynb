```python
class Demo:
  @classmethod
  def class_example_function(*args):
    return args

  @staticmethod
  def static_example_function(*args):
    return args

print(Demo.class_example_function("hi"))
print(Demo.static_example_function("hi"))
```

```python
class  Car(object):
  sell_price_ratio =  0.7
  def  __init__(self, details):
    self.__details = details

  def  info(self):
    print(f"id : {id(self)}")
    print(f'info : {self.__details}')

  def  get_buy_price(self):
    return  f'buy price : {self.__details.get("price")}'

  def  get_sell_price(self):
    return  f'sell price : {int(self.__details.get("price")*Car.sell_price_ratio)}'

  @classmethod
  def set_sell_price_price(cls, ratio):
    cls.sell_price_ratio = ratio

  @staticmethod
  def  is_luxury(instance):
    if instance.__details.get("company") ==  "bmw" or instance.__details.get("company") ==  "benz":
      return  True
    else:
      return False


car1 = Car({"company":  "bmw",  "price":  1000})
print(car1.get_buy_price())
print(car1.get_sell_price())
car1.sell_price_ratio = 0.5
car1.set_sell_price_price(0.5)
Car.set_sell_price_price(0.5)
print(car1.get_sell_price())
print(car1.is_luxury(car1))
print(Car.is_luxury(car1))
```
