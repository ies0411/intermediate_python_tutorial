```python
class ParentSample(object):
	def  __init__(self):
		self.value = 10

	def get_value(self):
		 return self.value

class ChildSample(ParentSample):
	def get_value(self):
		return self.value * 10

c = ChildSample()
c.get_value()
```

```python
class SampleOverloading:
    def __init__(self, num):
        self.num= num

    def __add__(self, other):
        self.num += other.num

    def __sub__(self, num):
        self.num -= num

n = SampleOverloading(10)
m = SampleOverloading(20)
n + m
print(n.num)
```

```python
import datetime

class  Logger(object):
	def  log(self, msg):
		print(msg)

class  TimestampLogger(Logger):
	def  log(self, msg):
		message =  f"{datetime.datetime.now()},{msg}"
		# super().log(message) 이렇게 해도됨
		super(TimestampLogger, self).log(message)


class  DateLogger(Logger):
	def  log(self, msg):
		message =  f"{datetime.datetime.now().strftime('%Y-%m-%d'),{msg}}"
		super(DateLogger, self).log(message)


l = Logger()
t = TimestampLogger()
d = DateLogger()

l.log("logger")
t.log("logger")
d.log("logger")
```
