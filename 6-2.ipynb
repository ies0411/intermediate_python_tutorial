```python
def sum_coroutine():
    total = 0
    while True:
        x = (yield total)
        total += x

c = sum_coroutine()
print(next(c))
print(c.send(1))
print(c.send(2))
print(c.send(3))
```

```python
def accumulate():
    total = 0
    while True:
        x = (yield)
        if x is None:
            return total
        total += x

def sum_coroutine():
    while True:
        total = yield from accumulate()
        print(total)

c = sum_coroutine()
print(next(c))

for i in range(1, 11):
    c.send(i)
c.send(None)
```

```python
from inspect import getgeneratorstate

def  co_task(n):
  tmp_n = yield n
  print(f'tmp_n : {tmp_n}')
  print("receive")
  z = yield n + tmp_n
  print(f'z : {z}')

c = co_task(5)
try:
  print(getgeneratorstate(c))
  print(next(c))
  print(getgeneratorstate(c))
  print(c.send(10))
  print(c.send(20))
except StopIteration:
	print("Coroutine ended.")
```
