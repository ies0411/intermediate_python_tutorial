```python
def count_letters(word):
    counter = {}
    for letter in word:
        if letter not in counter:
            counter[letter] = 0
        counter[letter] += 1
    return counter

count_letters("apple")
```

```python
def count_letters(word):
    counter = {}
    for letter in word:
        counter.setdefault(letter, 0)
        counter[letter] += 1
    return counter
count_letters("apple")
```

```python
from types import MappingProxyType

writable_data = {'a':1,"b":2}
read_only_data = MappingProxyType(writable_data)

print(read_only_data['a'])

writable_data['a'] = 3
print(read_only_data['a'])
```
