```python
import gc
import weakref

class Weakreference:
   def __init__(self, name):
     self.name = name
     self.f = None
   def __del__(self):
     self.f = None
     print(f"{self.name},{id(self)} is destroyed")

x = Weakreference("foo")
y = Weakreference("poo")
gc.enable()
x.f = weakref.ref(y)
y.f = weakref.ref(x)
del x
del y
```
