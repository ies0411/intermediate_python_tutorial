```python
from multiprocessing import Process, current_process, Value, Array
import os


def generate_update_number(v):
    for _ in range(50):
        v.value += 1


processes = []
share_value = Value("i", 0)
for _ in range(1, 11):
    p = Process(target=generate_update_number, args=(share_value,))
    processes.append(p)
    p.start()

for p in processes:
    p.join()

print(share_value.value)

```

```python
from multiprocessing import Process, Pipe, current_process
import time
import os


def task(id, end_number, conn):
    process_id = os.getpid()
    process_name = current_process().name
    total_number = 0
    for i in range(end_number):
        total_number += i
    print(f'total_number : {total_number}')
    conn.send(total_number)
    print(f'child_id : {process_id}')
    print(f'child name : {process_name}')
    print(f'id : {id}')

parent_process_id = os.getpid()
print(f'parent_process_id :{parent_process_id}')
parent_conn, child_conn = Pipe()
p = Process(name="child", target=task, args=(1, 10, child_conn))
p.start()
p.join()
print(parent_conn.recv())
```

```python
from multiprocessing import Process, Queue, current_process
import time
import os


def task(id, end_number, q):
    process_id = os.getpid()
    process_name = current_process().name
    print(f'child process_id : {process_id}')
    print(f'child process_name: {process_name}')
    sub_total = 0
    for i in range(end_number):
        sub_total += i
    q.put(sub_total)


parent_process_id = os.getpid()
print(f'parent_process_id : {parent_process_id}')

processes = []
q = Queue()

for i in range(5):
    t = Process(name=str(i), target=task, args=(i, 10, q))
    processes.append(t)
    t.start()

for process in processes:
    process.join()

q.put("finish")
total = 0
while True:
    tmp = q.get()
    if tmp == "finish":
        break
    else:
        total += tmp
print(total)

```
