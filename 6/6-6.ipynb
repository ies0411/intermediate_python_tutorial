```python
from multiprocessing import Process
import time

def task(name):
    print(name)
    time.sleep(3)


p = Process(target=task, args=("first",))
p.start()
# p.terminate()
print(p.is_alive())
p.join()
print(p.is_alive())
```

```python
from concurrent.futures import ProcessPoolExecutor
from multiprocessing import current_process

import os

def  task(args):
	name, end_number = args
	process_id = os.getpid()
	process_name = current_process().name
	print(f'process_id : {process_id}')
	print(f'process_name : {process_name}')
	print(f'name : {name}')
	print(f'end_number : {end_number}')
	total_number = 0
	for n in  range(end_number):
		total_number += n
	return total_number

numbers = [1,  2,  3,  4,  5]

with ProcessPoolExecutor()  as executor:
	results = executor.map(task,  [('task_'+str(i), number)  for i, number in  enumerate(numbers)])

for result in results:
	print("Result:", result)
```
