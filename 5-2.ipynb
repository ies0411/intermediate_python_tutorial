```python
import time


class Timer(object):
    def __init__(self, msg):
        self.__msg = msg

    def __enter__(self):
        self.__start = time.time()
        return self.__start

    def __exit__(self, exc_type, exc_value, exc_traceback):
        if exc_type:
            print(f"exception : {exc_type,exc_value,exc_traceback}")
        else:
            print(f"{self.__msg} : {time.time()-self.__start}")
        return True


with Timer("timer :") as v:
    time.sleep(1)

```

```python
import contextlib
import time


@contextlib.contextmanager
def file_manager(file_name, method):
    f = open(file_name, method)
    yield f
    f.close()


with file_manager("text.txt", "w") as f:
    f.write("context manager")

```

```python
import contextlib
import time

@contextlib.contextmanager
def Timer(msg):
    start_time = time.time()
    try:
        yield start_time  # __enter__
    except BaseException as e:
        print(f"{msg, e}")
        raise
    else:  # __exit__
        print(f"{msg} : {time.time()-start_time}")


with Timer("timer :") as v:
  time.sleep(1)
```
