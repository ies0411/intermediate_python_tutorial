```python
num_list = [x * 10  for x in  range(1,  11)]
def  ex_func(x):
	return x**2

result = list(map(ex_func, num_list))
print(result)
```

```python
num_list = [x * 10  for x in  range(1,  11)]

result = map(lambda i: i**2,num_list)
print(list(result))
```

```python
num_list = [x * 10  for x in  range(1,  11)]

def func(nums):
	def double_squre(x):
		return x**2
	return  map(double_squre, nums)

result = list(func(num_list))
print(result)
```

```python
def  fucn(nums):
	def  is_even(x):
		return x %  2  ==  0
	return  filter(is_even, nums)

num_list = [x for x in  range(1,  11)]

print(list(fucn(num_list)))
```

```python
from functools import  reduce

num_list = [x for x in  range(1, 11)]
result =  reduce(lambda  x, y: x + y, num_list)
print(result)
```

```python
from functools import  reduce

num_list = [x for x in  range(1, 11)]

def func(nums):
	def  custom_sum(x, y):
		return x + y
	return  reduce(custom_sum, nums)

print(func(num_list))
```
