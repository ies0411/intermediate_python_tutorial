```python
import logging
import threading
import time
total_list = [None] * 2
def thread_func(name, num, index):
  print(f"start : {name}")
  # time.sleep(1)
  total = 0
  for i in num:
    total += i
  print(f"finish : {name}")
  total_list[index] = total

x = threading.Thread(target=thread_func, args=("First",range(100),0,))
y = threading.Thread(target=thread_func, args=("Second",range(100),1,), daemon=True)

x.start()
y.start()


print(x.daemon)
print(y.daemon)

time.sleep(1)
x.join()
# y.join()

print(total_list[0])
print(total_list[1])

print("main finish")
```

```python
from concurrent.futures import ThreadPoolExecutor
import time
import os


def task(name,end_number):
    print(f'start : {name}')
    result = 0
    for i in range(end_number):
        result = result + i
    print(f'finish : {name}')
    return result


excutor = ThreadPoolExecutor(max_workers=os.cpu_count())
task1 = excutor.submit(task, "first", 100)
task2 = excutor.submit(task, "second", 10)

print(task1.result())
print(task2.result())
```

```python
from concurrent.futures import ThreadPoolExecutor
import time
import os


def task(args):
  name, end_number = args
  print(f'start : {name}')
  result = 0
  for i in range(end_number):
      result = result + i
  print(f'finish : {name}')
  return result


tasks = [("first", 100), ("second", 10)]

with ThreadPoolExecutor(max_workers=os.cpu_count()) as executor:
    results = executor.map(task, tasks)

for result in results:
  print(result)

```
