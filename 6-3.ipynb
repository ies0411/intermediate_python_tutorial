```python
import logging
import threading
import time

def thread_func(name, num):
  print(f"start : {name}")
  time.sleep(1)
  total = 0
  for i in num:
    total += i
  print(f"finish : {name}")
  return total

x = threading.Thread(target=thread_func, args=("First",range(100000),))
y = threading.Thread(target=thread_func, args=("Second",range(100),), )

x.start()
y.start()


print(x.daemon)
print(y.daemon)


x.join()
y.join()

print("main finish")
```
